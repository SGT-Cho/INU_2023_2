```python
def selectionSort(a, n):
    for i in range(1, n):
        smallest = i
        for j in range(i + 1, n + 1):
            if a[j] < a[smallest]:
                smallest = j
        a[i], a[smallest] = a[smallest], a[i]
            
        
        
def checkSort(a, n):
    isSorted = True
    for i in range(1, n):
        if (a[i] > a[i+1]):
            isSorted = False
        if (not isSorted):
            break
    if isSorted:
        print("정렬 완료")
    else:
        print("정렬 오류 발생")
```

```python
def bubbleSort(a,n):
    for i in range(n-1, 0, -1):
        for j in range(1, i + 1):
            if a[j] > a[j + 1]:
                a[j], a[j + 1] = a[j + 1], a[j]
```

```python
def insertionSort(a,n):
    n = len(a)
    for i in range(2, n):
        v = a[i]
        j = i
        while a[j - 1] > v and j > 1:
            a[j] = a[j - 1]
            j -= 1
        a[j] = v

```

```python
def shellSort(a,n):
    n=n+1
    h = 1
    while h < n-1 // 3:
        h = 3 * h + 1  # 첫 번째 h 값 계산
    
    while h > 0:
        for i in range(h + 1, n):
            v = a[i]
            j = i
            while j > h and a[j - h] > v:
                a[j] = a[j - h]
                j -= h
            a[j] = v
        h = h // 3
```
